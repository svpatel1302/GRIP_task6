```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
# Prepare the data data
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
dt_first = DecisionTreeClassifier()
dt_first.fit(X_train, y_train)
y_pred = dt_first.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt_first, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
score1 =classification_report(y_test, y_pred)
print(score1)
```

```python
# List of values to try for max_depth:
max_depth_range = list(range(1, 6))
# List to store the accuracy for each value of max_depth:
accuracy = []
for depth in max_depth_range:
    
    df = DecisionTreeClassifier(max_depth = depth, 
                             random_state = 0)
    df.fit(X_train, y_train)
    y_pred = df.predict(X_test)
    score = df.score(X_test, y_test)
    accuracy.append(score)
print(accuracy)
```

```python
from sklearn.model_selection import GridSearchCV
param_dist={
    "criterion":["gini","entropy"],
    "max_depth":[1,2,3,4,5]
}
classifier=DecisionTreeClassifier()
grid=GridSearchCV(classifier, param_grid=param_dist, cv=10, n_jobs=-1)
grid.fit(X_train,y_train)
GridSearchCV(cv=10, estimator=DecisionTreeClassifier(), n_jobs=-1,
             param_grid={'criterion': ['gini', 'entropy'],
                         'max_depth': [1, 2, 3, 4, 5]})
print(grid.best_params_) 
```

```python
grid_predictions = grid.predict(X_test) 
  # print classification report 
print(classification_report(y_test, grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions)) 
```
